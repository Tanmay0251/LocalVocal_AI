```python
import requests
import urllib.request
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
API_KEY = open('API_KEY').read()
SEARCH_ENGINE_ID = open('SEARCH_ENGINE_ID').read()
url = 'https://www.googleapis.com/customsearch/v1'
```

```python
search_query = 'colgate toothpaste'
```

```python
params = {
    'q': search_query,
    'key': API_KEY,
    'cx': SEARCH_ENGINE_ID,
    'searchType': 'image'
}
```

```python
response = requests.get(url, params=params)
```

```python
results = response.json()['items']
```

```python
print('Results:')
listofimages = []
for i in results:
    print(i['link'])
    urllib.request.urlretrieve(i['link'], 'temp.jpg')
    img = cv2.imread('temp.jpg')
    listofimages.append(img)    
```

# From this point we will show the images located in the links in jupyter notebook 

```python
print('Suggestions based on the search query')
for i in range(len(listofimages)):
    plt.axis('off')
    plt.imshow(listofimages[i])
    plt.show()
```
